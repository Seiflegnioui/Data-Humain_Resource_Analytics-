```python
def annotate_countplot(ax):
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(f'{int(height)}',
                        (p.get_x() + p.get_width() / 2, height),
                        ha='center', va='bottom',
                        fontsize=9, color='black',
                        xytext=(0, 3),
                        textcoords='offset points')

def annotate_stacked_bars(ax, df):
    for cnum, col in enumerate(df.columns):
        for xpos, val in enumerate(df[col]):
            if val > 0:
                bottom = df.iloc[xpos, :cnum].sum() if cnum > 0 else 0
                ax.text(xpos, bottom + val / 2, f'{val:.1f}%',
                        ha='center', va='center', fontsize=9, color='white')

counts = df.groupby(["Department", "JobSatisfaction"]).size().unstack(fill_value=0)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

counts_age = df.groupby(["Department", "AgeGroup"]).size().unstack(fill_value=0)
percentages_age = counts_age.div(counts_age.sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(2, 3, figsize=(29, 10))

fig.delaxes(axes[0, 0])
fig.delaxes(axes[1, 0])

sns.countplot(x='Department', hue='Attrition', data=df, ax=axes[0,1])
axes[0,1].set_title("Number of Employees in each Department")
axes[0,1].tick_params(axis='x', rotation=45)
annotate_countplot(axes[0,1])

sns.countplot(x='Department', hue='JobLevel', data=df, ax=axes[0,2])
axes[0,2].set_title("Job Levels per Department")
axes[0,2].tick_params(axis='x', rotation=45)
annotate_countplot(axes[0,2])

percentages.plot(kind='bar', stacked=True, colormap='viridis', ax=axes[1,1])
axes[1,1].set_title("Job Satisfaction % per Department")
axes[1,1].set_ylabel("Percentage (%)")
axes[1,1].legend(title="Job Satisfaction", bbox_to_anchor=(1.05, 1))
annotate_stacked_bars(axes[1,1], percentages)

percentages_age.plot(kind='bar', stacked=True, colormap='viridis', ax=axes[1,2])
axes[1,2].set_title("AgeGroup % per Department")
axes[1,2].set_ylabel("Percentage (%)")
axes[1,2].legend(title="AgeGroup", bbox_to_anchor=(1.05, 1))
annotate_stacked_bars(axes[1,2], percentages_age)

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df: pd.DataFrame, jobroles=None, outliers_columns=None):
    if (outliers_columns is None) or (jobroles is None):
        outliers_columns = ["MonthlyIncome", "HourlyRate", "DailyRate", "MonthlyRate"]
        jobroles = df["JobRole"].unique()
    
    for jobrole in jobroles:
        for col in outliers_columns:
            sliced_df = df[df["JobRole"] == jobrole]
            Q1 = sliced_df[col].quantile(0.25)
            Q3 = sliced_df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            mask = (df["JobRole"] == jobrole) & ((df[col] < lower_bound) | (df[col] > upper_bound))
            df = df[~mask] 
            print(f"Removed {mask.sum()} outliers in {jobrole} for column {col}")
    
    return df
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler

encoders = {}

prepared__df = remove_outliers(df, None, None)
def label_encoder(df: pd.DataFrame, columns):
    for col in columns:
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col].tolist())
        encoders[f"encoder-{col}"] = encoder  

columns = ["AgeGroup","Attrition","BusinessTravel","Gender","SalarySlab","OverTime"]
label_encoder(prepared__df, columns)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scalers = {}

def normalize(df, columns):
    for col in columns:
        normalizer = MinMaxScaler()
        df[col] = normalizer.fit_transform(df[[col]])
        scalers[f'scaler-{col}'] = normalizer
        
def standardize(df, columns):
    for col in columns:
        standr = StandardScaler()
        df[col] = standr.fit_transform(df[[col]])
        scalers[f'scaler-{col}'] = standr

normalize(prepared__df, [
    "AgeGroup","BusinessTravel","Education","EnvironmentSatisfaction","JobInvolvement","JobLevel",
    "JobSatisfaction","SalarySlab","PerformanceRating","RelationshipSatisfaction",
    "StockOptionLevel","WorkLifeBalance"
])

standardize(prepared__df, [
    "DailyRate","DistanceFromHome","HourlyRate","MonthlyIncome","MonthlyRate","NumCompaniesWorked",
    "PercentSalaryHike","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole",
    "YearsSinceLastPromotion","YearsWithCurrManager","LoyaltyIndex"
])
```

```python
from imblearn.combine import SMOTETomek
from collections import Counter
from sklearn.model_selection import train_test_split
```

```python
X = prepared__df.drop(columns=["EmployeeNumber","Attrition"])
Y = prepared__df["Attrition"]
```

```python
smote = SMOTETomek(random_state=42)
balanced_x,balanced_y = smote.fit_resample(X,Y)

X_train, X_test, y_train, y_test = train_test_split(
    balanced_x, balanced_y, test_size=0.2, random_state=42, stratify=balanced_y
)
X_test
```

```python
original_scaled = scalers["scaler-AgeGroup"].inverse_transform(X_train[["AgeGroup"]])

original_values = encoders["encoder-AgeGroup"].inverse_transform(original_scaled.astype(int).ravel())

print(len(original_values),len(X_train[["AgeGroup"]]))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

histories = {}
models = {
    "SVC": SVC(),
    "LinearSVC": LinearSVC(),
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
}

prev_acc = 0
attrition_classifier_model = None
best_model_name = None

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    test_accuracy = accuracy_score(y_test, y_pred)
    print(f'{name} -- Test Accuracy: {test_accuracy:.3f}')
    
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    mean_score = scores.mean()
    print(f"Cross-validation Accuracy: {mean_score:.3f}")
    
    if mean_score > prev_acc:
        prev_acc = mean_score
        attrition_classifier_model = model
        best_model_name = name
    
    histories[name] = {
        "Test-accuracy": test_accuracy,
        "Cross-validation accuracy": mean_score
    }

print("\nBest Model:", best_model_name)
print("Best CV Accuracy:", prev_acc)
print("\nAll results:", histories)
```

```python
df_hist =  pd.DataFrame(histories).T

df_hist.plot(kind='bar', figsize=(8, 5))
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0.8, 1.0)  
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```

```python
feature_importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': attrition_classifier_model.feature_importances_
})

feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
feature_importances['Importance (%)'] = feature_importances['Importance'] * 100

print(feature_importances.head(9))
```

```python
from sklearn.inspection import PartialDependenceDisplay
import matplotlib.pyplot as plt

y_pred_proba = attrition_classifier_model.predict_proba(X_train)[:, 1]
baseline = y_pred_proba.mean()

fig, ax = plt.subplots(2, 2, figsize=(14, 10))  

ax = ax.flatten()

ax[0].plot(feature_importances["Feature"].head(9), feature_importances["Importance (%)"].head(9), marker="o")
ax[0].set_xticklabels(feature_importances["Feature"].head(9), rotation=90)
ax[0].set_xlabel("Factor for attrition")
ax[0].set_ylabel("Percentage of importance")
ax[0].set_title("Feature Importance")

PartialDependenceDisplay.from_estimator(attrition_classifier_model, X=X_train, centered=True,
                                        ice_lines_kw={"color": "black"}, pd_line_kw={"color": "red"},
                                        features=["YearsWithCurrManager"], kind="both", ax=ax[1])
ax[1].set_title("PDP: YearsWithCurrManager")

PartialDependenceDisplay.from_estimator(attrition_classifier_model, X=X_train, centered=True,
                                        ice_lines_kw={"color": "black"}, pd_line_kw={"color": "red"},
                                        features=["DistanceFromHome"], kind="both", ax=ax[2])
ax[2].set_title("PDP: DistanceFromHome")

PartialDependenceDisplay.from_estimator(attrition_classifier_model, X=X_train, centered=True,
                                        ice_lines_kw={"color": "black"}, pd_line_kw={"color": "red"},
                                        features=["YearsAtCompany"], kind="both", ax=ax[3])
ax[3].set_title("PDP: YearsAtCompany")

plt.tight_layout()
plt.show()

print("Baseline (avg predicted probability):", baseline)
```

The average probability of attrition in our dataset is 0.50. After training our Random Forest classifier, we obtained the feature importances and Partial Dependence Plots (PDPs), which reveal the following insights:

- StockOptionLevel is the most influential factor, contributing 8.60% to the decision of attrition.

- YearsWithCurrManager comes next, contributing 5%. The PDP shows that as employees spend more years with the same manager, the probability of attrition decreases, compared to the average probability of 0.50.

- Job Involvement, Job Satisfaction, and Environment Satisfaction follow, with importance percentages ranging between 4.75% and 4.24%.

- YearsAtCompany also shows a decreasing probability of attrition as tenure increases.

- DistanceFromHome contributes 3.55%, and the ICE plots indicate that as the distance from home increases, the probability of attrition increases relative to the baseline.

Potential solution: Based on these insights, strategies to reduce attrition could include:

1 - Enhancing stock options or incentives to retain employees.

2 - Improving employee-manager relationships, e.g., mentorship programs, regular feedback, or recognition initiatives.

3 - Boosting job involvement, satisfaction, and work environment through engagement programs or flexible work arrangements.

4 - Considering employee commute factors when designing remote work or relocation support programs.
