```python
def annotate_countplot(ax):
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(f'{int(height)}',
                        (p.get_x() + p.get_width() / 2, height),
                        ha='center', va='bottom',
                        fontsize=9, color='black',
                        xytext=(0, 3),
                        textcoords='offset points')

def annotate_stacked_bars(ax, df):
    for cnum, col in enumerate(df.columns):
        for xpos, val in enumerate(df[col]):
            if val > 0:
                bottom = df.iloc[xpos, :cnum].sum() if cnum > 0 else 0
                ax.text(xpos, bottom + val / 2, f'{val:.1f}%',
                        ha='center', va='center', fontsize=9, color='white')

counts = df.groupby(["Department", "JobSatisfaction"]).size().unstack(fill_value=0)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

counts_age = df.groupby(["Department", "AgeGroup"]).size().unstack(fill_value=0)
percentages_age = counts_age.div(counts_age.sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(2, 3, figsize=(29, 10))

fig.delaxes(axes[0, 0])
fig.delaxes(axes[1, 0])

sns.countplot(x='Department', hue='Attrition', data=df, ax=axes[0,1])
axes[0,1].set_title("Number of Employees in each Department")
axes[0,1].tick_params(axis='x', rotation=45)
annotate_countplot(axes[0,1])

sns.countplot(x='Department', hue='JobLevel', data=df, ax=axes[0,2])
axes[0,2].set_title("Job Levels per Department")
axes[0,2].tick_params(axis='x', rotation=45)
annotate_countplot(axes[0,2])

percentages.plot(kind='bar', stacked=True, colormap='viridis', ax=axes[1,1])
axes[1,1].set_title("Job Satisfaction % per Department")
axes[1,1].set_ylabel("Percentage (%)")
axes[1,1].legend(title="Job Satisfaction", bbox_to_anchor=(1.05, 1))
annotate_stacked_bars(axes[1,1], percentages)

percentages_age.plot(kind='bar', stacked=True, colormap='viridis', ax=axes[1,2])
axes[1,2].set_title("AgeGroup % per Department")
axes[1,2].set_ylabel("Percentage (%)")
axes[1,2].legend(title="AgeGroup", bbox_to_anchor=(1.05, 1))
annotate_stacked_bars(axes[1,2], percentages_age)

plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.utils import resample

df_yes = df[df["Attrition"] == "Yes"]
df_no = df[df["Attrition"] == "No"]

min_length = min(len(df_yes), len(df_no))

df_yes_bal = df_yes.sample(n=min_length, random_state=42)
df_no_bal = df_no.sample(n=min_length, random_state=42)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
df_yes_bal['SalarySlab'].value_counts().sort_index().plot(
    kind='bar', color='skyblue', edgecolor='black')
plt.title("SalarySlab Distribution (Yes Attrition)")
plt.xlabel("Salary Slab")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
df_no_bal['SalarySlab'].value_counts().sort_index().plot(
    kind='bar', color='orange', edgecolor='black')
plt.title("SalarySlab Distribution (No Attrition)")
plt.xlabel("Salary Slab")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
print(len(df_no_bal),len(df_yes_bal))
```

```python
import numpy as np

attrition_balanced_df = pd.concat([df_yes_bal, df_no_bal], axis=0)
obs_table = pd.crosstab(attrition_balanced_df["Attrition"], attrition_balanced_df["SalarySlab"])
chi2,p,dof,exp_table = stats.chi2_contingency(obs_table)

chi2_critical = stats.chi2.ppf(1-0.05, df=dof)

print(chi2_critical,chi2)


# Values
values = [chi2_critical, chi2]
labels = ['Chi2 Critical Value', 'Chi2 Statistic']
colors = ['#a6cee3', '#1f78b4']  

plt.figure(figsize=(8, 3))
bars = plt.barh(labels, values, color=colors, edgecolor='black', height=0.6)

# Add values at end of bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.1, bar.get_y() + bar.get_height()/2, f'{width:.2f}', va='center', fontsize=12)

plt.xlim(0, max(values)*1.3)
plt.title('Chi-square Statistic vs Critical Value', fontsize=14)
plt.xlabel('Value')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

```

CHI-2 Test:

- H0 = the salary does not affect the attrition
- H1 = there a significant meaning of the salary to the attrition
- alpha is 0.05

The analysis revealed a statistically significant difference in the average distance from home between employees who left the company (attrited) and those who stayed, with a p-value of 0.0042.

On average, attrited employees lived approximately 1.67 km farther from work than non-attrited employees.

While this difference is statistically significant, it may not be practically meaningful, as the distance gap is relatively small and may not alone be a strong driver of attrition.
